```python
#permet d'effctuer une mise à l'échelle des données , de standariser les caractéristique en les centrant autour de 0
#en les mettant à l'échelle pour avoir une variance unique 
from sklearn.preprocessing import StandardScaler

# Création d'une instance de la classe StandardScaler
scaler = StandardScaler()
```

```python
import numpy as np 
# Utilisation de la méthode fit_transform() 
matrice_centree_reduite = scaler.fit_transform(data)
matrice_centree_reduite
```

```python
import numpy as np 
moyennes = np.mean(matrice_centree_reduite, axis=0)
#vérifier si les donneés sont centrée 
tolerance_moyenne = 1e-10 #Comme seuil 
colonnes_centrees = np.all(np.abs(moyennes) < tolerance_moyenne)
print("Est ce que les colonnes sont centrées :",colonnes_centrees)
```

```python
#vérifier si les donneés sont réduites 
ecarts_types = np.std(matrice_centree_reduite, axis=0)
ecarts_types
```

```python
import numpy as np
# Utilisation de la fonction np.corrcoef() pour calculer la matrice de corrélation
matrice_correlation = np.corrcoef(matrice_centree_reduite, rowvar=False)
```

```python
#Matrice symétrique avec une diagonale de 1 , elle représente la corélation entre les variables 
matrice_correlation
```

```python
# Utilisation de la fonction np.linalg.eig() pour calculer les valeurs propres et les vecteurs propres
valeurs_propres, vecteurs_propres= np.linalg.eig(matrice_correlation)
valeurs_propres
```

```python
# Calcul de l'inertie totale
inertie_totale = np.sum(valeurs_propres)
inertie_expliquee = (valeurs_propres / inertie_totale)
inertie_expliquee
```

```python
# Utilisation de la fonction np.dot() pour calculer les composantes principales
composantes_principales = np.dot(matrice_centree_reduite, vecteurs_propres)
composantes_principales
```

```python
import pandas as pd

vecteurs_pondere =-1* vecteurs_propres * np.sqrt(valeurs_propres)

data_frame = pd.DataFrame(vecteurs_pondere)
data_frame
```

```python
import matplotlib.pyplot as plt

# Tracer les vecteurs pondérés
plt.figure(figsize=(10, 10))
plt.scatter(vecteurs_pondere[:, 0], vecteurs_pondere[:, 1], s=50)

# Ajouter un cercle de rayon 1
cercle = plt.Circle((0, 0), 1, edgecolor='black', facecolor='none', linestyle='--', linewidth=2)
plt.gca().add_patch(cercle)

for i, (x, y) in enumerate(zip(vecteurs_pondere[:, 0], vecteurs_pondere[:, 1])):
    # Calcul de l'inertie expliquée pour la composante principale actuelle
    inertie_exp = inertie_expliquee[i]

    # Ajouter le texte avec l'inertie de la composante principale
    plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='red', width=0.002)
    
    # Ajouter le nom de la variable correspondante
    nom_variable = data.columns[i]
    plt.text(x, y, nom_variable, fontsize=8, verticalalignment='center', horizontalalignment='center')

plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
plt.axvline(0, color='black', linestyle='--', linewidth=0.5)
plt.title('Cercle de Corrélation ')
plt.xlabel('Axe 1')
plt.ylabel('Axe  2')
plt.grid(True)
plt.show()
```

```python
#Calcul des coordonnées au carré pour chaque individu le long de chaque composante principale
squared_coordinates = composantes_principales**2
squared_coordinates
```

```python
#Somme des coordonnées au carré pour chaque individu sur l'ensemble des composantes principales
total_squared_coordinates = np.sum(squared_coordinates, axis=1)
```

```python
#Calcul de la qualité de représentation pour chaque individu le long de chaque composante principale
representation_quality = squared_coordinates / total_squared_coordinates[:, np.newaxis]
representation_quality
```

```python
#Affichage de la qualité de représentation pour chaque individu le long de chaque composante principale
print("\nQualité de représentation pour chaque individu :")
print(representation_quality[:, :2])
```

```python
#Visualisation de la qualité de représentation à l'aide d'un graphique à barres avec des pourcentages
num_individuals = representation_quality[:, :2].shape[0]
num_components = representation_quality[:, :2].shape[1]

fig, ax = plt.subplots(figsize=(10, 6))
#Conversion en pourcentages
representation_percentage = representation_quality[:, :2] * 100

for i in range(num_components):
    ax.bar(np.arange(num_individuals) + i * 0.2, representation_percentage[:, i], width=0.2, label=f'AXE {i + 1}')

ax.set_xlabel('Individuals')
ax.set_ylabel('Quality of Representation (%)')
ax.legend()
plt.show()
```
