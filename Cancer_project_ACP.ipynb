```python
from sklearn.preprocessing import StandardScaler

# CrÃ©ation d'une instance de la classe StandardScaler
scaler = StandardScaler()
# La classe StandardScaler est utilisÃ©e pour effectuer une mise Ã  l'Ã©chelle standard sur les donnÃ©es.
# Cela implique de centrer les donnÃ©es autour de zÃ©ro et de les mettre Ã  la mÃªme Ã©chelle en ajustant
# la moyenne et l'Ã©cart type de chaque variable.
# Cette Ã©tape est souvent utilisÃ©epour normaliser les donnÃ©es avant de les utiliser dans un modÃ¨le d'apprentissage automatique.

# Utilisation de la mÃ©thode fit_transform() pour effectuer une mise Ã  l'Ã©chelle standard
matrice_centree_reduite = scaler.fit_transform(data)
matrice_centree_reduite
```

```python
import numpy as np
# Utilisation de la fonction np.corrcoef() pour calculer la matrice de corrÃ©lation
matrice_correlation = np.corrcoef(matrice_centree_reduite, rowvar=False)
matrice_correlation
```

```python
# Utilisation de la fonction np.linalg.eig() pour calculer les valeurs propres et les vecteurs propres
valeurs_propres, vecteurs_propres= np.linalg.eig(matrice_correlation)
valeurs_propres
```

```python
# Calcul de l'inertie totale
inertie_totale = np.sum(valeurs_propres)
inertie_expliquee = (valeurs_propres / inertie_totale)
inertie_expliquee
```

```python
# Utilisation de la fonction np.dot() pour calculer les composantes principales
composantes_principales = np.dot(matrice_centree_reduite, vecteurs_propres)
composantes_principales
```

```python
import pandas as pd

vecteurs_pondere =-1* vecteurs_propres * np.sqrt(valeurs_propres)

data_frame = pd.DataFrame(vecteurs_pondere)
data_frame
```
